```python
from typing import Optional
from typing import List
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
class Solution:
    # 83 删除排序链表中的重复元素
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        left=head
        right=head
        while right:
            right=right.next
            while right and left.val == right.val:
                right=right.next
            left.next=right
            left=left.next
        return head
    def deleteDuplicates_1(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return head
        cur=head
        while cur.next:
            if cur.val == cur.next.val:
                cur.next=cur.next.next
            else:
                cur=cur.next
        return head
    
    # 82 删除排序链表中的重复元素2
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        dummy_head=ListNode(-101)
        dummy_head.next=head
        cur=dummy_head
        flag=0
        while cur.next:
            while cur.next.next and cur.next.val == cur.next.next.val:
                cur.next=cur.next.next
                flag=1
            if flag==1: # 去掉重复的
                cur.next=cur.next.next
                flag=2 # 不许更新cur
            if flag==0: # 干干净净的
                cur=cur.next
            flag=0

        return dummy_head.next
    
    def deleteDuplicates_2(self, head: Optional[ListNode]) -> Optional[ListNode]:
        dummy_head=ListNode(-1)
        dummy_head.next=head
        
        cur=dummy_head
        while cur.next and cur.next.next:
            if cur.next.val ==cur.next.next.val:
                temp=cur.next
                while temp and temp.next and temp.val == temp.next.val:
                    temp=temp.next
                cur.next=temp.next
            else:
                cur=cur.next
        return dummy_head.next
    
    # 206 反转链表
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return head
        
        cur=head
        lis=[]
        while cur:
            lis.append(cur.val)
            cur=cur.next
        dummy_head=ListNode(0)
        curr=dummy_head
        for num in reversed(lis):
            cur_node=ListNode(num)
            curr.next=cur_node
            curr=curr.next
        return dummy_head.next
    def reverseList_iter(self, head: Optional[ListNode]) -> Optional[ListNode]:
        pre=None
        cur=head
        while cur:
            next=cur.next
            cur.next=pre
            pre=cur
            cur=next
        return pre
    def reverseList_rec(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        new_head=self.reverseList(head.next)
        head.next.next=head
        head.next=None
        return new_head
        
    # 92 反转链表2
    def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        if left==1:
            pre=None
            cur=head
            left_node=head
            for _ in range(right):
                next=cur.next
                cur.next=pre
                pre=cur
                cur=next
            left_node.next=cur
            return pre
        if left >= 2:
            left2_node=head
            for _ in range(left-2):
                left2_node=left2_node.next
            pre=None
            left_node=left2_node.next
            cur=left2_node.next
            for _ in range(right-left+1):
                next=cur.next
                cur.next=pre
                pre=cur
                cur=next
            left2_node.next=pre
            left_node.next=cur
            return head
    # 25 K个一组翻转链表
    def reverseKGroup_mine(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        begin,end=[],[]
        cur=head
        if not head or not head.next or k==1:
            return head
        while cur:
            begin.append(cur)
            begin_node=cur
            count=1
            while cur and count<k:
                cur=cur.next
                count+=1
            if count==k and cur is not None:
                end.append(cur)
                pre=None
                cur=begin_node
                for _ in range(k):
                    next=cur.next
                    cur.next=pre
                    pre=cur
                    cur=next
            
            else: # 不成比例的结束（end比begin少一个）
                n=len(end)
                if n==0:
                    return begin[0]
                elif n==1:
                    begin_node1=begin[0]
                    begin_node2=begin[1]
                    begin_node1.next=begin_node2
                    return end[0]
                else:
                    for i in range(n-1):
                        begin_node=begin[i]
                        end_node=end[i+1]
                        begin_node.next=end_node
                    begin_node1=begin[-2]
                    begin_node2=begin[-1]
                    begin_node1.next=begin_node2
                    return end[0]
        # 成比例结束
        n=len(end)
        if n==1:
            return end[0]
        for i in range(n-1):
            begin_node=begin[i]
            end_node=end[i+1]
            begin_node.next=end_node
        return end[0]
    
    def reverse(self,head,tail):
        pre=head
        cur=pre.next
        first=cur
        while cur != tail:
            next = cur.next
            cur.next=pre
            pre=cur
            cur=next
        head.next=pre
        first.next=tail
        return first
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        dummy_head=ListNode(0)
        dummy_head.next=head
        cur=dummy_head
        tail=dummy_head.next
        index=0
        while tail:
            index +=1
            if index % k ==0:
                cur=self.reverse(cur,tail.next)
                tail=cur.next
            else:
                tail=tail.next
        return dummy_head.next
    
    # 328 奇偶链表
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        odd,even=head,head.next
        even_begin=even
        cur=head.next
        step=1
        while cur:
            cur=cur.next
            if step % 2 ==1 and cur is not None:
                odd.next=cur
                odd=odd.next
            else:
                even.next=cur
                even=even.next
            step+=1
        odd.next=even_begin
        return head
    
    # 234 回文链表
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        left=head
        right=head.next
        while right and right.next:
            left=left.next
            right=right.next.next
        cur=left.next
        pre=None
        while cur:
            next=cur.next
            cur.next=pre
            pre=cur
            cur=next
        right=pre
        left=head
        while right:
            if left.val != right.val:
                return False
            left=left.next
            right=right.next
        return True
    
    # 138 随机链表的复制
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        if not head:
            return None
        node_dict=dict()
        curr=head
        while curr:
            new_node=Node(curr.val,None,None)
            node_dict[curr]=new_node
            curr=curr.next
        curr=head
        while curr:
            if curr.next:
                node_dict[curr].next=node_dict[curr.next]
            if curr.random:
                node_dict[curr].random=node_dict[curr.random]
            curr=curr.next
        return node_dict[head]
    
    # 61 旋转链表
    def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if not head or k==0:
            return head
        cur=head
        n=1
        while cur.next:
            cur=cur.next
            n+=1
        cur.next=head
        k=n - k % n
        cur=head
        for _ in range(k-1):
            cur=cur.next
        ans=cur.next
        cur.next=None
        return ans

    # 148 排序链表
    def merge(self, head1,head2):
        dummy_head=ListNode(-1)
        cur=dummy_head
        left,right=head1,head2
        while left and right:
            if left.val <= right.val:
                cur.next=left
                cur=cur.next
                left=left.next
            else:
                cur.next=right
                cur=cur.next
                right=right.next
        if left:
            cur.next=left
        else:
            cur.next=right
        return dummy_head.next
    
    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        left,right=head,head.next
        while right and right.next:
            left=left.next
            right=right.next.next
        left_head=head
        right_head=left.next
        left.next=None
        return self.merge(self.sortList(left_head),self.sortList(right_head))
    
    # 23 合并K个升序链表
    def mergeKLists_bad(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        k=len(lists)
        cur=None
        for i in range(k):
            if lists[i]==None:
                continue
            if cur==None:
                cur=lists[i]
            else:
                cur=self.merge(cur,lists[i])
        return cur
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        if not lists:
            return None
        size=len(lists)
        return self.merge_sort(lists,0,size-1)
    
    def merge_sort(self,lists,left,right):
        if left==right:
            return lists[left]
        mid=left + (right - left) // 2
        node_left = self.merge_sort(lists,left,mid)
        node_right = self.merge_sort(lists,mid+1,right)
        return self.merge(node_left,node_right)
    
    # 141 环形链表
    def hasCycle_hash(self, head: Optional[ListNode]) -> bool:
        hash_list=set()
        cur=head
        while cur:
            if cur in hash_list:
                return True
            hash_list.add(cur)
            cur=cur.next
            
        return False
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        slow,fast=head,head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
            if slow==fast:
                return True
        return False
    
    # 142 环形链表2
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow,fast=head,head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
            if slow==fast:
                slow=head
                while slow!=fast:
                    slow=slow.next
                    fast=fast.next
                return slow
        return None
    
    # 160 相交链表
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        m,n=1,1
        curA,curB=headA,headB
        while curA.next:
            curA=curA.next
            m+=1
        while curB.next:
            curB=curB.next
            n+=1
        if curA != curB:
            return None
        curA,curB=headA,headB
        if m >= n:
            for _ in range(m-n):
                curA=curA.next
        else:
            for _ in range(n-m):
                curB=curB.next
        while curA != curB:
            curA=curA.next
            curB=curB.next
        return curA
    
    def getIntersectionNode_m_n(self, headA: ListNode, headB: ListNode) -> ListNode:
        if headA == None or headB == None:
            return None
        pA = headA
        pB = headB
        while pA != pB:
            pA = pA.next if pA != None else headB
            pB = pB.next if pB != None else headA
        return pA
    
    # LCR140 训练计划2
    def trainingPlan(self, head: Optional[ListNode], cnt: int) -> Optional[ListNode]:
        slow,fast=head,head
        for _ in range(cnt):
            fast=fast.next
        while fast:
            slow=slow.next
            fast=fast.next
        return slow
    
    # 143 重排链表
    def reorderList_list(self, head: Optional[ListNode]) -> None:
        cur=head
        n=0
        node_lists=[]
        while cur:
            n+=1
            node_lists.append(cur)
            cur=cur.next
        if n<=1:
            return head
        for i in range(n//2):
            node_lists[i].next=node_lists[n-1-i]
            node_lists[n-1-i].next=node_lists[i+1]
        node_lists[n//2+1].next=None
        return node_lists[0]
        
    def reorderList(self, head: ListNode) -> None:
        if not head or not head.next:
            return head
        # 先找反转点
        slow,fast=head,head.next
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
        # 反转后面的列表
        cur=slow.next
        slow.next=None
        pre=None
        while cur:
            next=cur.next
            cur.next=pre
            pre=cur
            cur=next
        # 拼接
        slow,fast=head,pre
        while fast:
            temp_slow=slow.next
            temp_fast=fast.next
            fast.next=temp_slow
            slow.next=fast
            slow=temp_slow
            fast=temp_fast

        return head
    
    #2 两数相加
    def reverse_list(self,head):
        pre=None
        cur=head
        while cur:
            next=cur.next
            cur.next=pre
            pre=cur
            cur=next
        return pre
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy_head=ListNode(0)
        cur=dummy_head
        temp_up=0
        while l1 and l2:
            temp=l1.val+l2.val+temp_up
            temp_val=temp % 10
            temp_up=temp//10
            cur.next=ListNode(temp_val)
            cur=cur.next
            l1=l1.next
            l2=l2.next
        while l1:
            temp=l1.val+temp_up
            temp_val=temp % 10
            temp_up=temp//10
            cur.next=ListNode(temp_val)
            cur=cur.next
            l1=l1.next
        while l2:
            temp=l2.val+temp_up
            temp_val=temp % 10
            temp_up=temp//10
            cur.next=ListNode(temp_val)
            cur=cur.next
            l2=l2.next
        if temp_up==1:
            cur.next=ListNode(1)
        return dummy_head.next
    # 445 两数相加2
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        l1=self.reverse_list(l1)
        l2=self.reverse_list(l2)
        carry=0
        dummy_head=ListNode(0)
        cur=dummy_head
        while l1 or l2 or carry:
            if l1:
                num1=l1.val
                l1=l1.next
            else:
                num1=0
            if l2:
                num2=l2.val
                l2=l2.next
            else:
                num2=0
            cur.next=ListNode((num1+num2+carry)%10)
            cur=cur.next
            carry=(num1+num2+carry)//10
        return self.reverse_list(dummy_head.next)

                
s=Solution()
s.reverseKGroup([1,2,3,4,5],2)                   
                
            
                
                
        
        
                
            
    
        
```
