```python
from typing import List
class Solution:
    """ 611 有效三角形的个数 """
    def triangleNumber(self, nums: List[int]) -> int:
        nums.sort()
        ans=0
        n=len(nums)
        for i in range(2,n): # 遍历最大边
            left,right=0,i-1
            while left<right:
                if nums[left]+nums[right]<=nums[i]:
                    left+=1
                else:
                    ans+=(right-left)
                    right-=1
        return ans
    
    """ 15 三数之和 """
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        ans=[]
        nums.sort()
        n=len(nums)
        for i in range(2,n):
            # 最大值也得去去重
            if (nums[i]<0) or (i<n-1 and nums[i]==nums[i+1]):
                continue
            left,right=0,i-1
            while left<right:  
                if nums[left]+nums[right]==-nums[i]:
                    ans.append([nums[left],nums[right],nums[i]])
                    left+=1
                    right-=1
                    # 去掉重复值
                    while left<right and nums[left] == nums[left-1]:
                        left+=1
                    while left<right and nums[right] == nums[right+1]:
                        right-=1
                while left<right and nums[left]+nums[right]>-nums[i]:
                    right-=1
                while left<right and nums[left]+nums[right]<-nums[i]:
                    left+=1
        return ans
    """ 16 最接近的三数之和 """
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        ans=float('inf')
        n=len(nums)
        nums.sort()
        for i in range(2,n):
            if i<n-1 and nums[i]==nums[i+1]:
                continue
            left,right=0,i-1
            while left<right:  
                total=nums[left]+nums[right]+nums[i]
                if total==target:
                    return target
                if abs(total-target)<abs(ans-target):
                    ans=total
                
                if total < target:
                    left+=1
                else:
                    right-=1   
        return ans
    
    def threeSumClosest_best(self, nums: List[int], target: int) -> int:
        ans=float('inf')
        n=len(nums)
        nums.sort()
        for i in range(2,n):
            if i<n-1 and nums[i]==nums[i+1]:
                continue
            
            # 判断最大值是否≤target
            total=nums[i]+nums[i-1]+nums[i-2]
            if total <= target:
                if abs(total-target)<abs(ans-target):
                    ans=total
                continue
            # 判断最小值是否≥target
            total=nums[0]+nums[1]+nums[i]
            if total >= target:
                if abs(total-target)<abs(ans-target):
                    ans=total
                continue
            
            left,right=0,i-1
            while left<right:  
                total=nums[left]+nums[right]+nums[i]
                if total==target:
                    return target
                if abs(total-target)<abs(ans-target):
                    ans=total
                
                if total < target:
                    left+=1
                else:
                    right-=1   
        return ans
        
            
        
            
s=Solution()
nums = [0,1,2]
s.threeSumClosest_best(nums,0)        
```
