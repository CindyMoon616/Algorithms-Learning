一、axis使用教程:

sum/mean/max/argmax 这类“聚合/归约”函数：消掉axis  
对于二维矩阵而言：axis=0就是对列进行处理；axis=1就是对行进行处理

np.concatenate(arrays, axis=k)：在第 k 维把长度加起来(其它维度必须相同)

np.stack(arrays, axis=k)：多出一维，axis 指新维插在第几维位置(形状必须相同)

<img src="pics/2.png" width="35%" />

```python

import numpy as np

array_23=np.array([[1,3,4],[2,4,5]])
array_22=np.array([[10,30],[20,40]])
print(array_23.shape,array_23)
print(array_22.shape,array_22)
axis_1=np.sum(array_23,axis=1)
print("sum",axis_1.shape,axis_1)

axis_1=np.argmax(array_23,axis=1)
print("argmax",axis_1.shape,axis_1)

axis_1=np.concatenate([array_23,array_22],axis=1)
print("concatenate",axis_1.shape,axis_1)

array_23_0=np.array([[10,30,40],[20,40,50]])
axis_1=np.stack([array_23,array_23_0,array_23,array_23_0],axis=1)
print("stack",axis_1.shape,axis_1)
```

```python

import numpy as np

array_423=np.array([[[1,3,4],[2,4,5]],[[12,33,42],[22,41,52]],[[221,223,334],[222,224,335]],[[2221,1113,1114],[1112,1114,1115]]])
print(array_423)
print(array_423.shape)
axis_1=np.sum(array_423,axis=1)
print("sum",axis_1.shape,axis_1)

axis_1=np.argmax(array_423,axis=1)
print("argmax",axis_1.shape,axis_1)

axis_1=np.concatenate([array_423],axis=1)
print("concatenate",axis_1.shape,axis_1)
```

$H(p,g)=\sum_{i}p_i\times log q_i$  
其中p为真实分布，q为模型分布，−logq去惩罚“真实应该高概率的部分”（模型分布越好，logq越高，-logq越低）
