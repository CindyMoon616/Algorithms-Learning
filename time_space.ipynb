```python
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        I_need={}
        for i_num,i in enumerate(nums):
            if i in I_need:
                return [i_num,I_need[i]]
            I_need[target-i]=i_num
solution=Solution()
print(solution.twoSum([2,27,11,7,4,22,15],9))
```

二、空间复杂度  
1、空间复杂度为$O(1)$的例子（一般忽略）:  
创造变量、修改变量

2、空间复杂度为$O(n)$的例子：  
定义一个长度为n的数组、字典、链表

3、空间复杂度为$O(n^2)$的例子：  

```python
# O(m+n)
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        m=len(matrix)
        n=len(matrix[0])
        for i in range(m):
            for j in range(n):
                if matrix[i][j]==0:
                    for k in range(n):
                        matrix[i][k]="not 0" if matrix[i][k]!=0 else matrix[i][k]
                    for k in range(m):
                        matrix[k][j]="not 0" if matrix[k][j]!=0 else matrix[k][j]
        for i in range(m):
            for j in range(n):
                matrix[i][j]=0 if matrix[i][j]=="not 0" else matrix[i][j]
        return matrix

solution=Solution()
matrix=[[0,1,2,0],[3,4,5,2],[1,3,1,5]]
print(solution.setZeroes(matrix))
```
