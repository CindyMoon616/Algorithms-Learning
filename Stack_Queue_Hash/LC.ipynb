```python
from typing import List
class Solution:
    # 1047 删除字符串中的所有相邻重复项
    def removeDuplicates(self, s: str) -> str:
        stack=[0]
        for ch in s:
            if ch == stack[-1]:
                stack.pop()
            else:
                stack.append(ch)
        return "".join(stack[1:])

    # 20 有效的括号
    def isValid(self, s: str) -> bool:
        stack=[]
        isvalid={")":"(","]":"[","}":"{"}
        for ch in s:
            if stack and stack[-1] == isvalid.get(ch,0):
                stack.pop()
            else:
                stack.append(ch)
        if not stack:
            return True
        else:
            return False

    # 224 基本计算器
    def calculate(self, s: str) -> int:
        stack=[]
        cnt=None
        not_nums=["(",")","+","-"]
        for ch in s:
            if ch == ")":
                cur=stack.pop()
                cnt=0
                temp=None
                while cur != "(":
                    if cur =="-":
                        cnt-=temp
                        temp=None
                    elif cur == "+":
                        cnt+=temp
                        temp=None
                    else:
                        temp=int(cur)
                    cur=stack.pop()
                if temp: # （后面是数字，就没有加入cnt
                    cnt+=temp
                stack.append(cnt)
            elif ch in not_nums:
                stack.append(ch)
            elif ch != " ":
                val=int(ch)
                if stack and stack[-1] not in not_nums:
                    stack[-1]=stack[-1]*10+val
                else:
                    stack.append(val)
        if stack: # 只剩下没有括号的了
            temp=cnt
            ans=0
            while stack:
                cur=stack.pop()
                if cur == "-":
                    ans-=temp
                    temp=None
                elif cur == "+":
                    ans+=temp
                    temp=None
                else:
                    temp=cur
            if temp:
                ans+=temp
            return ans
        return cnt

    # 227 基本计算器2
    def calculate2(self, s: str) -> int:
        stack=[]
        flag=None
        pre=None
        next=None
        not_nums=["+","-","*","/"]
        for ch in s:
            if ch == " ":
                continue
            if ch.isdigit():
                if stack and stack[-1] not in not_nums:
                    stack[-1]=stack[-1]*10+int(ch)
                else:
                    stack.append(int(ch))
                continue
            if flag == "*" or flag == "/":
                next=stack.pop()
                temp=pre * next if flag == "*" else pre // next 
                stack.append(temp)
                flag=None
            if ch == "*" or ch =="/":
                pre=stack.pop()
                flag = ch
            if ch == "+" or ch == "-":
                stack.append(ch)
        if flag == "*" or flag == "/":
            next=stack.pop()
            temp=pre * next if flag == "*" else pre // next 
            stack.append(temp)

        # 现在只剩加减法了
        if stack: 
            temp=None
            ans=0
            while stack:
                cur=stack.pop()
                if cur == "-":
                    ans-=temp
                    temp=None
                elif cur == "+":
                    ans+=temp
                    temp=None
                else:
                    temp=cur
            if temp:
                ans+=temp
            return ans
    # 394 字符串编码
    def decodeString(self, s: str) -> str:
        stack=[]
        times=1 # 表示
        n=len(s)
        index=0
        while index < n:
            ch=s[index]
            if ch.isdigit():
                times=int(ch)
                index+=1
                while index < n and s[index].isdigit():
                    ch=s[index]
                    times=times*10+int(ch)
                    index+=1
                stack.append(times)
            elif ch != "]":
                if stack and isinstance(stack[-1],int):
                    stack.append(ch)
                else:
                    stack.append(1)
                    stack.append(ch)
                index+=1
            else:
                cur=stack.pop()
                temp=""
                while cur != "[":
                    times=stack.pop()
                    temp=cur*times+temp
                    cur=stack.pop()
                times=stack.pop()
                temp=times*temp
                stack.append(1)
                stack.append(temp)
                index+=1
        # 现在只有 数字+字母
        ans=""
        while stack:
            ch=stack.pop()
            times=stack.pop()
            ans=times*ch+ans
        return ans
        
    # 最长有效括号
    def longestValidParentheses(self, s: str) -> int:
        stack=[-1]
        ans=0
        for i in range(len(s)):
            if s[i]=="(":
                stack.append(i)
            else:
                stack.pop()
                if stack:
                    ans=max(ans,i-stack[-1])
                else:
                    stack.append(i)
        return ans
    
    # 739 每日温度
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        



                



    


s=Solution()
print(s.decodeString("abc3[cd]xyz"))

    
# 155 最小栈
class MinStack:
    def __init__(self):
        self.stack=[]

    def push(self, val: int) -> None:
        if not self.stack:
            self.stack.append((val,val))
        else:
            min_val=min(val,self.stack[-1][-1])
            self.stack.append((val,min_val))

    def pop(self) -> None:
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1][0]
    
    def getMin(self) -> int:
        return self.stack[-1][-1]

# LCR125 图书整理2        
class CQueue:
    def __init__(self):
        self.inStack=[]
        self.outStack=[]

    def appendTail(self, value: int) -> None:
        self.inStack.append(value)

    def deleteHead(self) -> int:
        if len(self.outStack) == 0 and len(self.inStack)==0:
            return -1
        if len(self.outStack) == 0:
            while len(self.inStack)>0:
                self.outStack.append(self.inStack.pop())
        top=self.outStack.pop()
        return top
```
