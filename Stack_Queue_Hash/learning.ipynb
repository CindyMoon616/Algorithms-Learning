```python
# 顺序栈：采用一段连续的存储空间（如数组）依次存放从栈底到栈顶的元素，并通过指针𝑡𝑜𝑝标记当前栈顶元素在数组中的位置
# 局限性：当栈满或需要扩容时，往往需要移动大量元素，效率较低
class Stack_list:
    def __init__(self,size=100):
        self.stack=[]
        self.size=size
        self.top=-1 # -1代表空栈
    
    def is_empty(self):
        return self.top==-1

    def is_full(self):
        return self.top+1==self.size
    
    # 入栈
    def push(self,value):
        if self.is_full():
            raise Exception("栈已满")
        self.stack.append(value)
        self.top+=1
    
    # 出栈
    def pop(self):
        if self.is_empty():
            raise Exception("栈为空")
        value=self.stack.pop()
        self.top+=1
        return value
    
    # 查看栈顶元素
    def peek(self):
        if self.is_empty():
            raise Exception("栈为空")
        return self.stack[self.top]


# 链式栈：采用单链表实现，每次新元素都插入到链表头部，𝑡𝑜𝑝始终指向链表的头节点，即栈顶元素的位置。
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None

class Stack:
    def __init__(self):
        self.top=None

    def is_empty(self):
        return self.top is None
    
    # 入栈
    def push(self,value):
        new_node=Node(value)
        new_node.next=self.top
        self.top=new_node

    # 出栈
    def pop(self):
        if self.is_empty():
            raise Exception("栈为空")
        value=self.top.value
        self.top=self.top.next
        return value

    # 查看栈顶元素
    def peek(self):
        if self.is_empty():
            raise Exception("栈为空")
        return self.top.value
```

```python
# 单调栈：要求从栈顶到栈底的元素单调递增或单调递减
# 496 下一个更大元素
from typing import List
def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:
    res=[]
    stack=[]
    num_map=dict()
    for num in nums2:
        while stack and num>stack[-1]:
            num_map[stack[-1]]=num
            stack.pop()
        stack.append(num)
    for num in nums1:
        res.append(num_map.get(num,-1))
    return res
```

# 队列
1、队列：先进先出，只允许在一端插入元素，在另一端删除元素  
2、顺序存储队列：用数组构建（一般循环移动）  
（1）入队：𝑠𝑒𝑙𝑓.𝑟𝑒𝑎𝑟 = (𝑠𝑒𝑙𝑓.𝑟𝑒𝑎𝑟 +1) mod 𝑠𝑒𝑙𝑓.𝑠𝑖𝑧e  
（2）出队：𝑠𝑒𝑙𝑓.𝑓𝑟𝑜𝑛𝑡 = (𝑠𝑒𝑙𝑓.𝑓𝑟𝑜𝑛𝑡+1) mod 𝑠𝑒𝑙𝑓.𝑠𝑖𝑧e  
（3）特意空出一个位置，约定「队头指针在队尾指针的下一位置」为队满：  
队满：(𝑠𝑒𝑙𝑓.𝑟𝑒𝑎𝑟+1) mod 𝑠𝑒𝑙𝑓.𝑠𝑖𝑧𝑒 == 𝑠𝑒𝑙𝑓.𝑓𝑟𝑜𝑛t  
队空：𝑠𝑒𝑙𝑓.𝑓𝑟𝑜𝑛𝑡 == 𝑠𝑒𝑙𝑓.𝑟𝑒𝑎r  
3、链式存储队列：单链表  
（0）哨兵节点：self.front=Node(0) 不动如山
（1）入队：self.rear后移  
（2）出队：self.front.next后移，注意如果如果出队后队列为空，即self.rear == self.front.next，此时将self.rear = self.front  
4、优先队列：是一种为每个元素分配优先级的特殊队列结构。每次访问或移除元素时，总是优先处理优先级最高的元素  
（1）数组（顺序存储）：入队时直接将元素插入数组末尾，时间复杂度为𝑂(1)；出队时需遍历整个数组以找到优先级最高的元素并删除，时间复杂度为𝑂(𝑛)。
（2）链表（链式存储）：链表内元素按优先级有序排列，入队时需找到合适插入位置，时间复杂度为𝑂(𝑛)；出队时直接移除链表头节点，时间复杂度为𝑂(1)。
（3）二叉堆结构：通过二叉堆维护优先级顺序，入队操作（插入新元素）和出队操作（弹出优先级最高元素）均为𝑂(log𝑛)，效率较高。  
5、双向队列：一种线性表数据结构，允许在队列的两端进行插入和删除操作，既可以从队头入队/出队，也可以从队尾入队/出队。（用链式存储） 
（1）节点结构：每个节点包含两个指针域prev和next  
（2）头尾指针：维护指向队头节点和队尾节点的指针  
（3）哨兵节点：简化边界处理（头尾双哨兵）

```python
# 优先队列
class Heapq:
    # 堆调整方法：将以index为根的子树调整为大顶堆
    def heapAdjust(self,nums:list,index:int,end:int):
        # nums:堆数组  index：当前需要调整的根节点下标  end：堆的最后一个元素下标
        while 2*index+1 <= end:
            left=index*2+1
            right=left+1
            if right <= end and nums[right]>nums[left]:
                larger=right
            else:
                larger=left
            if nums[larger] > nums[index]:
                nums[index], nums[larger]= nums[larger], nums[index]
            else:   # 如果根节点已经是最大值
                break
            index=larger

    # 建堆：将数组整体调整为大顶堆
    def heapify(self,nums:list):
        size = len(nums)
        # 从最后一个非叶子节点开始，依次向前调整：
        for i in range((size-2)//2,-1,-1):
            self.heapAdjust(nums,i,size-1)
    
    # 入队操作：插入到新元素到堆中
    def heappush(self,nums:list,value):
        nums.append(value) # 先加到末尾
        i=len(nums)-1
        # 自下而上调整
        while i>0:
            parent=(i-1)//2
            if nums[parent] >= value:
                break
            nums[i]=nums[parent]
            i=parent
        nums[i]=value

    # 出队操作：弹出堆顶元素
    def heappop(self,nums:list):
        size=len(nums)
        if size==0:
            raise IndexError("heappop from empty heap")
        # 交换堆顶和末尾元素
        nums[0],nums[-1]=nums[-1],nums[0]
        top=nums.pop()
        if size>1:
            self.heapAdjust(nums,0,size-2)
        return top

    # 堆排序：原地将数组升序排序
    def heapSort(self,nums):
        self.heapify(nums)
        size=len(nums)
        #依次将堆顶元素（最大值）交换到末尾，缩小堆范围
        for i in range(size-1,0,-1):
            nums[0],nums[i]=nums[i],nums[0]
            self.heapAdjust(nums,0,i-1)
        return nums 

            
```

# 哈希表
1、哈希表的思想：通过哈希函数将键𝑘𝑒𝑦映射到表的某个区块，实现高效的数据插入与查找  
2、哈希函数：将元素的关键字映射为哈希表存储位置的函数（允许不同key映射到同一个区域）  
（1）直接定址法：直接取关键字本身或其线性函数作为哈希地址，即𝐻𝑎𝑠ℎ(𝑘𝑒𝑦) = 𝑘𝑒𝑦或H𝑎𝑠ℎ(𝑘𝑒𝑦) = 𝑎×𝑘𝑒𝑦 +𝑏，其中𝑎、𝑏为常数。（计算简单不会冲突、但如果key系数则会空间浪费）  
（2）保留余数法：设哈希表长度为m，选取不大于m的质数p，用𝐻𝑎𝑠ℎ(𝑘𝑒𝑦)=𝑘𝑒𝑦 mod 𝑝计算哈希地址。（最常用，减少冲突）  
（3）平方取中法：：先对关键字平方，再取中间若干位作为哈希地址。例如𝐻𝑎𝑠ℎ(𝑘𝑒𝑦)=(𝑘𝑒𝑦×𝑘𝑒𝑦)//100 mod 1000，即先平方，去掉末尾两位，再取中间三位。（使哈希地址分布更均匀，减少冲突）  
（4）基数转换法：将关键字视为某一进制数，转换为另一进制后，选取部分位作为哈希地址  
3、哈希冲突：不同key经过同一个哈希函数后得到相同的哈希地址  
（1）开放地址法：当哈希冲突发生时，通过探查哈希表中的其他「空地址」来存放冲突元素，直到找到空位为止。  
例如：𝐻(𝑖) = (𝐻𝑎𝑠ℎ(𝑘𝑒𝑦)+𝐹(𝑖)) mod 𝑚，其中𝑖=1,2,3,...,𝑛，𝑛 ≤ 𝑚−1。  
𝐹(𝑖)可以是线形探查i（每次向后找一位），或者二次探查（二次方步长向前后）或者伪随机探查  
（2）链地址法：：将哈希地址相同的元素以链表的形式存储在同一个槽位中。
