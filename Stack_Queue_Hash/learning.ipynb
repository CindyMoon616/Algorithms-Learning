```python
# é¡ºåºæ ˆï¼šé‡‡ç”¨ä¸€æ®µè¿žç»­çš„å­˜å‚¨ç©ºé—´ï¼ˆå¦‚æ•°ç»„ï¼‰ä¾æ¬¡å­˜æ”¾ä»Žæ ˆåº•åˆ°æ ˆé¡¶çš„å…ƒç´ ï¼Œå¹¶é€šè¿‡æŒ‡é’ˆð‘¡ð‘œð‘æ ‡è®°å½“å‰æ ˆé¡¶å…ƒç´ åœ¨æ•°ç»„ä¸­çš„ä½ç½®
# å±€é™æ€§ï¼šå½“æ ˆæ»¡æˆ–éœ€è¦æ‰©å®¹æ—¶ï¼Œå¾€å¾€éœ€è¦ç§»åŠ¨å¤§é‡å…ƒç´ ï¼Œæ•ˆçŽ‡è¾ƒä½Ž
class Stack_list:
    def __init__(self,size=100):
        self.stack=[]
        self.size=size
        self.top=-1 # -1ä»£è¡¨ç©ºæ ˆ
    
    def is_empty(self):
        return self.top==-1

    def is_full(self):
        return self.top+1==self.size
    
    # å…¥æ ˆ
    def push(self,value):
        if self.is_full():
            raise Exception("æ ˆå·²æ»¡")
        self.stack.append(value)
        self.top+=1
    
    # å‡ºæ ˆ
    def pop(self):
        if self.is_empty():
            raise Exception("æ ˆä¸ºç©º")
        value=self.stack.pop()
        self.top+=1
        return value
    
    # æŸ¥çœ‹æ ˆé¡¶å…ƒç´ 
    def peek(self):
        if self.is_empty():
            raise Exception("æ ˆä¸ºç©º")
        return self.stack[self.top]


# é“¾å¼æ ˆï¼šé‡‡ç”¨å•é“¾è¡¨å®žçŽ°ï¼Œæ¯æ¬¡æ–°å…ƒç´ éƒ½æ’å…¥åˆ°é“¾è¡¨å¤´éƒ¨ï¼Œð‘¡ð‘œð‘å§‹ç»ˆæŒ‡å‘é“¾è¡¨çš„å¤´èŠ‚ç‚¹ï¼Œå³æ ˆé¡¶å…ƒç´ çš„ä½ç½®ã€‚
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None

class Stack:
    def __init__(self):
        self.top=None

    def is_empty(self):
        return self.top is None
    
    # å…¥æ ˆ
    def push(self,value):
        new_node=Node(value)
        new_node.next=self.top
        self.top=new_node

    # å‡ºæ ˆ
    def pop(self):
        if self.is_empty():
            raise Exception("æ ˆä¸ºç©º")
        value=self.top.value
        self.top=self.top.next
        return value

    # æŸ¥çœ‹æ ˆé¡¶å…ƒç´ 
    def peek(self):
        if self.is_empty():
            raise Exception("æ ˆä¸ºç©º")
        return self.top.value
```

```python
# å•è°ƒæ ˆï¼šè¦æ±‚ä»Žæ ˆé¡¶åˆ°æ ˆåº•çš„å…ƒç´ å•è°ƒé€’å¢žæˆ–å•è°ƒé€’å‡
# 496 ä¸‹ä¸€ä¸ªæ›´å¤§å…ƒç´ 
from typing import List
def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:
    res=[]
    stack=[]
    num_map=dict()
    for num in nums2:
        while stack and num>stack[-1]:
            num_map[stack[-1]]=num
            stack.pop()
        stack.append(num)
    for num in nums1:
        res.append(num_map.get(num,-1))
    return res
```
