```python
import collections
from typing import List,Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def __init__(self):
        self.ans=-float(int)
    # 144 前序遍历
    def preorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        ans=[]
        def preorder(root):
            if not root:
                return
            ans.append(root.val)
            preorder(root.left)
            preorder(root.right)
        preorder(root)
        return ans

    def preorderTraversal_Stack(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        stack=[root]
        ans=[]
        while stack:
            node=stack.pop()
            ans.append(node.val)
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
        return ans
    
    # 94 中序遍历
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        ans=[]
        def inorder(root):
            if not root:
                return
            inorder(root.left)
            ans.append(root.val)
            inorder(root.right)
        inorder(root)
        return ans

    def inorderTraversal_Stack(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        stack=[]
        ans=[]
        cur=root
        while cur or stack:
            while cur:
                stack.append(cur)
                cur=cur.left
            # 到达最左节点
            cur=stack.pop()
            ans.append(cur.val)
            cur=cur.right
        return ans
    
    # 102 层序遍历
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        import collections
        queue=collections.deque([root])
        ans=[]
        while queue:
            cur_queue=[]
            n=len(queue)
            for _ in range(n):
                node=queue.popleft()
                cur_queue.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            ans.append(cur_queue)
        return ans
    
    # 103 锯齿型层序遍历
    def zigzagLevelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        import collections
        queue=collections.deque([root])
        ans=[]
        flag=0
        while queue:
            cur_queue=[]
            n=len(queue)
            for _ in range(n):
                node=queue.popleft()
                cur_queue.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            ans.append(cur_queue) if flag % 2 ==0 else ans.append(cur_queue[::-1])
            flag+=1
        return ans
    
    # 236 最近公共祖先
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if root==p or root==q:
            return root
        
        if root:
            node_left=self.lowestCommonAncestor(root.left,p,q)
            node_right=self.lowestCommonAncestor(root.right,p,q)
            if node_left and node_right:
                return root
            elif not node_left:
                return node_right
            else:
                return node_left
        return None
    
    def lowestCommonAncestor_Stack(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        stack=[root]
        parent={root:None}

        while p not in parent or q not in parent:
            node=stack.pop()
            if node.left:
                parent[node.left]=node
                stack.append(node.left)
            if node.right:
                parent[node.right]=node
                stack.append(node.right)
        
        ancestors=set()
        while p:
            ancestors.add(p)
            p=parent[p]
        
        while q not in ancestors:
            q=parent[q]
        
        return q
    
    # 104 二叉树的最大深度
    def maxDepth_queue(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        ans=0
        queue=collections.deque([root])
        while queue:
            ans+=1
            n=len(queue)
            for _ in range(n):
                node=queue.popleft()
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
        return ans
    
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        return 1+max(self.maxDepth(root.left),self.maxDepth(root.right))
    
    # 112 路径总和
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if not root:
            return False
        if not root.left and not root.right:
            return root.val == targetSum
        return self.hasPathSum(root.left,targetSum-root.val) or self.hasPathSum(root.right,targetSum-root.val)
    
    # 113 路径总和2
    def pathSum2(self, root: Optional[TreeNode], targetSum: int) -> List[List[int]]:
        ans=[]
        cur_list=[]
        def sum(root,targetSum):
            if not root:
                return
            cur_list.append(root.val)
            if not root.left and not root.right and root.val==targetSum:
                ans.append(cur_list[:])
            sum(root.left,targetSum-root.val)
            sum(root.right,targetSum-root.val)
            cur_list.pop()
        sum(root,targetSum)
        return ans
    
    # 101 对称二叉树
    def isSymmetric_queue(self, root: Optional[TreeNode]) -> bool:
        # 队列里存放的是“需要比较的一对儿”
        queue=collections.deque([root.left, root.right])
        while queue:
            # 每次取两个PK
            left_node = queue.popleft()
            right_node = queue.popleft()
            # 两个都空
            if not left_node and not right_node:
                continue
            if (left_node and not right_node) or (not left_node and right_node) or (left_node.val != right_node.val):
                return False
            # 将后代成对加入队列（注意顺序！）
            # 外侧对比：左的左 vs 右的右
            queue.append(left_node.left)
            queue.append(right_node.right)
            
            # 内侧对比：左的右 vs 右的左
            queue.append(left_node.right)
            queue.append(right_node.left)
        return True
    
    def isSymmetric(self, root: TreeNode) -> bool:
        if root == None:
            return True
        def check(left,right):
            if left==None and right==None:
                return True
            elif left==None and right!=None:
                return False
            elif left!=None and right==None:
                return False
            elif left.val != right.val:
                return False
            return check(left.left, right.right) and check(left.right,right.left)
        return check(root.left, root.right)
    
    # 124 二叉树的最大路径和
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        def max_tree(root):
            if not root:
                return 0
            max_left=max(max_tree(root.left),0)
            max_right=max(max_tree(root.right),0)
            cur_max=root.val+ max_left + max_right
            self.ans=max(cur_max,self.ans) # 这个是不连父结点的情况，可以直接当做答案
            return max(max(max_left,max_right) + root.val,0)
        max_tree(root)
        return self.ans
    

    # 199 二叉树的右视图
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        ans=[]
        queue=collections.deque([root])
        while queue:
            curr=[]
            n=len(queue)
            for _ in range(n):
                node=queue.popleft()
                curr.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            ans.append(curr[-1])
        return ans
    
    # 226 翻转二叉树
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        def invert(curr):
            if not curr:
                return
            temp=curr.left
            curr.left=curr.right
            curr.right=temp
            invert(curr.left)
            invert(curr.right)

        invert(root)
        return root

    
    
                


        

    



s=Solution()
```

```python
import collections
from typing import List,Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    # 105 二叉树还原
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        def build(preorder,inorder,n):
            if n==0:
                return
            k=0
            root_val=preorder[0]
            while inorder[k] != root_val:
                k+=1
            root=TreeNode(root_val)
            root.left=build(preorder[1:k+1],inorder[0:k],k)
            root.right=build(preorder[k+1:],inorder[k+1:],n-k-1)
            return root
        return build(preorder,inorder,len(preorder))

    def buildTree_hash(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        inorder_map = {val: idx for idx, val in enumerate(inorder)}
        def build(preorder_left, preorder_right, inorder_left, inorder_right):
            if preorder_left > preorder_right:
                return
            root_val=preorder[preorder_left]
            k=inorder_map[root_val]
            root=TreeNode(root_val)

            left_size=k-inorder_left # 左边有这么多个节点

            root.left=build(preorder_left+1, preorder_left+left_size, inorder_left, k-1)
            root.right=build(preorder_left+left_size+1,preorder_right,k+1,inorder_right)
            return root
        return build(0, len(preorder) - 1, 0, len(inorder) - 1)
    
    # 98 验证二叉搜索树
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        def validate(node, lower, upper):
            if not node:
                return True
            if node.val <= lower or node.val >= upper:
                return False
            # 左边节点＜自己，右边节点＞自己，其余边界条件不变
            return validate(node.left, lower, node.val) and validate(node.right, node.val, upper)
        return validate(root, float('-inf'), float('inf'))
    
    # 110 平衡二叉树
    def isBalanced(self, root: Optional[TreeNode]) -> bool:
        def height(node):
            if not node:
                return 0
            left_height=height(node.left)
            if left_height == -1:
                return -1
            right_height=height(node.right)
            if right_height == -1:
                return -1
            if abs(left_height-right_height) > 1:
                return -1
            return max(left_height,right_height)+1
        return height(root) != -1            
    
    # 128 最长连续序列
    def longestConsecutive(self, nums: List[int]) -> int:
        ans=0
        nums_set=set(nums)
        for num in nums_set:
            if num-1 in nums_set:
                continue
            cnt=0
            cur=num
            while cur in nums_set:
                cnt+=1
                cur+=1
            ans=max(ans,cnt)
        return ans
```
