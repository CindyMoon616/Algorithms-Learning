# 一、树与二叉树基础
1、树的定义：由𝑛 ≥ 0个节点及其相互之间的关系组成的有限集合。当𝑛=0时称为空树，n>0时称为非空树，具有以下特性：  
（1）仅有一个没有前驱的节点：根节点root  
（2）除根节点外，其余每个节点都有且仅有一个直接前驱节点：父节点  
（3）每个节点可以有0个或多个后继节点：子节点  
（4）当n>1时，除根节点以外的其余节点可分为m（>0）个互不相交的集合$T_i$，每个集合本身又是一棵树，称为根的子树

# 二、二叉树
1、二叉树的定义：是一种有序树，其中每个节点的度最多为2。每个节点的两个分支分别称为左子树和右子树，且左右子树的顺序不可交换。  
2、特殊二叉树：  
（1）满二叉树：指所有非叶子节点均有左右两个子节点，且所有叶子节点都集中在同一层的二叉树  
（2）完全二叉树：除了最后一层外，每一层的节点数都达到最大，且最后一层的所有节点都连续排列在最左侧。  
（3）二叉搜索树：左子树所有节点的值均小于该节点的值，则右子树所有节点的值均大于该节点的值  
<img src="../pics/14.png" width="50%" />  
（4）平衡二叉搜索树：是任意节点的左右子树高度差的绝对值不超过1，能够保证插入、查找和删除操作的时间复杂度均为𝑂(log𝑛)  
3、二叉树的顺序存储结构：  
（1）使用数组存储  
（2）若非叶子节点下标为$i$，则其左右孩子下标为$2i+1,2i+2$  
（3）若非根节点下标为$i$，则其父节点下标为$(i-1)//2$  
（4）适合完全二叉树（尤其是满二叉树），如果存在大量空节点，会造成空间浪费。

# 三、二叉树的遍历
1、二叉树的遍历定义：指从根节点出发，按照特定顺序依次访问二叉树中的所有节点，确保每个节点被且仅被访问一次。  
2、二叉树前序遍历：如果二叉树非空，则依次执行————访问当前根节点、递归遍历左子树、递归遍历右子树    
<img src="../pics/16.png" width="50%" />  
3、二叉树中序遍历：如果二叉树非空，则依次执行————递归遍历左子树、访问当前根节点、递归遍历右子树  
<img src="../pics/17.png" width="50%" />  
4、二叉树后序遍历：如果二叉树非空，则依次执行————递归遍历左子树、递归遍历右子树、访问当前根节点  
<img src="../pics/18.png" width="50%" />  
5、二叉树层序遍历：如果二叉树非空，则依次执行————先访问第1层、再访问第2层、直到访问最底层所有节点

```python
from typing import List
class TreeNode:
    def __init__(self,val=0,left=None,right=None):
        self.val=val
        self.left=left
        self.right=right




class Solution():
    # 前序遍历（递归）
    def preorderTraversal(self, root: TreeNode)-> List[int]:
        res=[]
        def preorder(node):
            if not node:
                return
            res.append(node.val) # 访问根节点
            preorder(node.left) # 递归遍历左子树
            preorder(node.right) # 递归遍历右子树
        preorder(root)
        return res
    # 前序遍历（栈）
    def preorderTraversal_Stack(self, root: TreeNode)-> List[int]:
        if not root:
            return []
        res=[]
        stack=[root]
        while stack:
            node=stack.pop()
            res.append(node.val)
            # 右子树先入栈，才会左边先出
            if node.right: 
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
        return res
    
    # 中序遍历（递归）
    def inorderTraversal(self, root: TreeNode)-> List[int]:
        res=[]
        def inorder(node):
            if not node:
                return
            inorder(node.left)
            res.append(node.val)
            inorder(node.right)
        inorder(root)
        return res
    # 中序遍历（栈）
    def inorderTraversal_Stack(self, root: TreeNode)-> List[int]:
        res=[]
        stack=[]
        cur=root

        while cur or stack: # 只要当前节点不为空就继续深入左边，只要当前Stack不为空就继续输出
            while cur:
                stack.append(cur)
                cur=cur.left
            # 到达最左边（注意，以后的cur全部是右节点了）
            node = stack.pop()
            res.append(node.val)
            cur=node.right # 转向右子树，继续上述过程（如果没有右子树，就自动去父节点实现中序）
        return res
    
    # 后序遍历（递归）
    def postorderTraversal(self, root: TreeNode)-> List[int]:
        res=[]
        def postorder(node):
            if not node:
                return
            postorder(node.left)
            postorder(node.right)
            res.append(node.val)
        postorder(root)
        return res
    
    # 后序遍历（栈）
    def postorderTraversal_Stack(self, root: TreeNode)-> List[int]:
        res=[]
        stack=[]
        prev = None # 记录上一个访问节点，判断右子树是否已访问
        while root or stack: # 只要节点不为空就深入左边，只要Stack不为空就继续输出
            while root:
                stack.append(root)
                root=root.left
            node=stack.pop() # 最左边的子节点了
            # 没有右子树 或者 右子树之前访问过（来第二次了）
            if not node.right or node.right == prev:
                res.append(node.val)
                prev=node
                root=None # 跳过while root
            else:
                # 右子树还未访问，当前父节点重新入栈，转而遍历右子树
                stack.append(node)
                root=node.right
        return res
    
    # 层序遍历
    def levelOrder(self, root: TreeNode)-> List[List[int]]:
        if not root:
            return []
        from collections import deque
        queue=deque([root]) # 初始化队列，根节点入队
        order = []

        while queue:
            level=[] # 当前层的节点值
            size=len(queue) # 当前层的节点数
            for _ in range(size):
                curr=queue.popleft()
                level.append(curr.val)
                if curr.left:
                    queue.append(curr.left)
                if curr.right:
                    queue.append(curr.right)
            if level:
                order.append(level)
            return order

            


```

# 四、二叉树的还原
1、二叉树的还原：通过已知的二叉树遍历序列，重建出原始的二叉树结构  
2、前序+中序：前序序列确定根节点，中序序列确定左右子树的范围。递归分割子序列，可以唯一还原原二叉树。  
3、中序+后序：后序序列确定根节点，中序序列确定左右子树的范围，方法与前序+中序类似，也能唯一还原二叉树。  
4、中序+层序：通过层序遍历确定每个子树的根节点，再结合中序遍历分割左右子树，也可以唯一还原二叉树。  
5、前序+后序：因为缺少中序信息，无法区分左右子树的分界。（满二叉树可唯一确定）

```python
from typing import List
class TreeNode:
    def __init__(self,val=0,left=None,right=None):
        self.val=val
        self.left=left
        self.right=right

class Solution:
    # 前序+中序
    def buildTree_12(self, preorder: List[int], inorder: List[int])-> TreeNode:
        def createTree(preorder, inorder, n):
            if n==0:
                return None # 当前子树节点数
            k=0
            while preorder[0]!= inorder[k]:
                k+=1
            # 找到inorder中根节点位置了
            node=TreeNode(inorder[k])
            # 递归构建左子树和右子树
            node.left=createTree(preorder[1:k+1],inorder[0:k],k)
            node.right=createTree(preorder[k+1:],inorder[k+1:],n-1-k)
            return node
        return createTree(preorder,inorder,len(inorder))
    
    # 中序和后序
    def buildTree_23(self, inorder: List[int], postorder: List[int])-> TreeNode:
        def createTree(inorder, postorder, n):
            if n==0:
                return None
            k=0
            root_val=postorder[n-1]
            while inorder[k] != root_val:
                k+=1
            node=TreeNode(root_val)
            node.left=createTree(inorder[:k],postorder[:k],k)
            node.right=createTree(inorder[k+1:],postorder[k:-1],n-1-k)
            return node
        return createTree(inorder,postorder,len(postorder))
    
    # 前序和后序
    def constructFromPrePost(self,preorder:List[int], postorder:List[int]) -> TreeNode:
        def createTree(preorder, postorder, n):
            if n==0:
                return None
            root_val=preorder[0]
            node=TreeNode(root_val)
            if n==1:
                return node
            # 前序遍历的第二个元素为左子树的根节点
            left_root_val=preorder[1]
            # 在后序遍历中查找左子树根节点的位置
            k=0
            while postorder[k] != left_root_val:
                k+=1
            node.left = createTree(preorder[1:k+2],postorder[0:k+1],k+1)
            node.right=createTree(preorder[k+2:],postorder[k+1:n-1],n-k-1)
            return node
        return createTree(preorder, postorder, len(preorder))
            
```

# 五、二叉搜索树
1、二叉搜索树：左子树所有节点的值均小于该节点的值，右子树所有节点的值均大于该节点的值，任意节点的左右子树也都分别是二叉搜索树。  
<img src="../pics/21.png" width="50%" />  
2、二叉搜索树的查找：va < root.val则递归查找左子树  
<img src="../pics/22.png" width="50%" />  
3、二叉搜索树的插入：在二叉搜索树中插入一个值为𝑣𝑎𝑙的节点（假设当前树中不存在𝑣𝑎𝑙）  
4、二叉搜索树的创建：根据给定数组中的元素，依次插入，构建出一棵二叉搜索树  
5、二叉树的删除：在二叉搜索树中删除值为val的节点





```python
from typing import List
class TreeNode:
    def __init__(self,val=0,left=None,right=None):
        self.val=val
        self.left=left
        self.right=right

class solution:
    # 二叉搜索树的查找
    def searchBST(self,root:TreeNode,val:int) -> TreeNode:
        if not root:
            return None
        if val == root.val:
            return root
        elif val < root.val:
            return self.searchBST(root.left,val)
        else:
            return self.searchBST(root.right,val)
        
    # 二叉搜索树的插入
    def insertIntoBST(self,root:TreeNode, val:int) -> TreeNode:
        if not root:
            return TreeNode(val)
        if val < root.val:
            root.left = self.insertIntoBST(root.left,val)
        elif val > root.val:
            root.right = self.insertIntoBST(root.right,val)
        # 如果val ==root.val，不插入（不允许重复），直接返回原树
        return root
    
    # 二叉搜索树的构建
    def buildBST(self,nums) -> TreeNode:
        root=None
        for num in nums:
            root=self.insertIntoBST(root,num)
        return root
    
    # 二叉搜索树的删除
    def deleteNode(self, root: TreeNode, val: int)-> TreeNode:
        if not root:
            return None
        if val<root.val:
            root.left=self.deleteNode(root.left,val)
            return root
        elif val>root.val:
            root.right=self.deleteNode(root.right,val)
            return root
        else:
            # 找到目标节点了
            if not root.left:
                # 情况1：左子树为空，直接返回右子树
                return root.right
            elif not root.right:
                # 情况2：右子树为空，直接返回左子树
                return root.left
            else:
                # 情况3：用右子树的最左节点（即后继节点）替换当前节点，并且递归删除后继节点
                successor = root.right
                while successor.left:
                    successor=successor.left
                root.val=successor.val
                root.right=self.deleteNode(root.right,successor.val) # 现在没有左子树和右子树，直接删除就行
                return root
```

# 五、并查集
1、并查集的定义：用于处理若干不相交集合的合并与查询操作。合并（Union）：将两个不同的集合合并为一个集合，查找（Find）：确定某个元素属于哪个集合，通常返回该集合的「代表元素」。  
2、用数组实现：将每个元素的集合编号设为其自身的下标，即每个元素自成一个集合。将一个集合中的所有元素的𝑖𝑑修改为另一个集合的𝑖𝑑，从而实现集合的合并  
3、基于森林实现：每个节点都指向其父节点


```python
class UnionFind:
    def __init__(self,n):
        self.fa=[i for i in range(n)]

    def find(self,x):
        while self.fa[x] !=x:
            # 将 x 的父节点直接指向其祖父节点，实现隔代压缩
            self.fa[x] = self.fa[self.fa[x]]
            '''# 或者直接连接根节点
            self.fa[x] = self.find(self.fa[x])'''
            x=self.fa[x]
        return x
    
    def union(self,x,y):
        root_x=self.find(x)
        root_y=self.find(y)
        if root_x == root_y:
            return False
        self.fa[root_x]=root_y
        return True
    
    def is_connected(self,x,y):
        return self.find(x) == self.find(y)
    
```
