```python
from typing import List
class Solution:
    # 3 无重复字符的最长子串
    def lengthOfLongestSubstring(self, s: str) -> int:
        n=len(s)
        if n<=1:
            return n
        str_set=set()
        ans=0
        left,right=0,1
        str_set.add(s[left])
        cnt=1
        while right<n:
            while s[right] in str_set:
                str_set.remove(s[left])
                left+=1
                cnt-=1
            str_set.add(s[right])
            right+=1
            cnt+=1
            ans=max(cnt,ans)
        return ans
    # 5 最长回文子串
    def longestPalindrome(self, s: str) -> str:
        n=len(s)
        max_ans=1
        ans=s[0]
        for i in range(0,n):
            # 奇数子串
            dist=1
            while i-dist>=0 and i+dist<=n-1:
                if s[i-dist] == s[i+dist]:
                    dist+=1
                else:
                    break
            if 2*dist-1>max_ans:
                max_ans=2*dist-1
                ans=s[i+1-dist:i+dist]

            # 偶数子串
            if i+1<=n-1 and s[i] == s[i+1]:
                dist=1
                while i-dist>=0 and i+1+dist<=n-1:
                    if s[i-dist] == s[i+1+dist]:
                        dist+=1
                    else:
                        break
                
                if 2*dist>max_ans:
                    max_ans=2*dist
                    ans=s[i+1-dist:i+1+dist]
        return ans
    
    # 415 字符串相加
    def addStrings(self, num1: str, num2: str) -> str:
        n1,n2=len(num1),len(num2)
        if n1 > n2:
            return self.addStrings(num2,num1)
        carry=0
        ans=[]
        for i in range(-1,-n1-1,-1):
            cnt=int(num1[i])+int(num2[i])+carry
            carry=cnt//10
            #ans+=(cnt%10)*10**(-i-1)
            ans.append(str(cnt%10))
        if n1 == n2:
            if carry:
                ans.append(str(carry))
            return "".join(reversed(ans))
        for i in range(-n1-1,-n2-1,-1):
            cnt=int(num2[i])+carry
            carry=cnt//10
            #ans+=(cnt%10)*10**(-i-1)
            ans.append(str(cnt%10))
        if carry:
            ans.append(str(carry))
        return "".join(reversed(ans))
    
    # 151 反转字符串中的单词
    def reverseWords(self, s: str) -> str:
        words=[]
        ans=[]
        for ch in reversed(s):
            if ch != " ":
                words.append(ch)
            elif words:
                if ans:
                    ans.append(" ")
                while words:
                    ans.append(words.pop())
        if words:
            if ans:
                ans.append(" ")
            while words:
                ans.append(words.pop())
        return("".join(ans))
    
    # 43 字符串相乘
    def multiply(self, num1: str, num2: str) -> str:
        if num1 == "0" or num2 == "0":
            return "0"
        len1, len2 = len(num1), len(num2)
        nums = [0 for _ in range(len1 + len2)]
        for i in range(len1-1,-1,-1):
            digit1 = int(num1[i])
            for j in range(len2-1,-1,-1):
                digit2 = int(num2[j])
                nums[i+j+1]+=digit1*digit2
        
        for i in range(len1+len2-1,0,-1):
            nums[i-1]+=nums[i]//10
            nums[i]%=10
        if nums[0]==0:
            return "".join([str(digit) for digit in nums[1:]])
        else:
            return "".join([str(digit) for digit in nums[:]])
        
    # 14 最长公公前缀
    def longestCommonPrefix(self, strs: List[str]) -> str:
        n=len(strs)
        length=[len(x) for x in strs]
        ans=[]
        for i in range(length[0]):
            ch=strs[0][i]
            for j in range(1,n):
                if i>=length[j] or strs[j][i] != ch:
                    return "".join(ans)
            ans.append(ch)
        return "".join(ans)
            
        

    


s=Solution()
num1 = "1"
num2 = "9"
strs = ["flower"]
s.longestCommonPrefix(strs)
```
