1ã€KMPç®—æ³•æ ¸å¿ƒæ€æƒ³ï¼šåœ¨å­—ç¬¦ä¸²åŒ¹é…è¿‡ç¨‹ä¸­ï¼Œå½“æ–‡æœ¬ä¸²ğ‘‡çš„æŸä¸ªå­—ç¬¦ä¸æ¨¡å¼ä¸²ğ‘å‘ç”Ÿä¸åŒ¹é…æ—¶ï¼Œå……åˆ†åˆ©ç”¨å·²åŒ¹é…çš„å‰ç¼€ä¿¡æ¯ï¼Œé€šè¿‡é¢„å¤„ç†å¾—åˆ°çš„ã€Œéƒ¨åˆ†åŒ¹é…è¡¨ã€ï¼ˆå³nextæ•°ç»„ï¼‰ï¼Œé¿å…æ–‡æœ¬æŒ‡é’ˆçš„å›é€€ï¼Œä»è€Œé«˜æ•ˆåœ°å‡å°‘ä¸å¿…è¦çš„æ¯”è¾ƒæ¬¡æ•°ï¼Œå®ç°å¿«é€ŸåŒ¹é…ã€‚

```python
# ç”Ÿæˆnextæ•°ç»„
# next[j] ï¼šå­ä¸²p[0:j+1]çš„æœ€é•¿ç›¸ç­‰å‰åç¼€çš„é•¿åº¦
def generateNext(p:str):
    m=len(p)
    next=[0 for _ in range(m)]
    left=0 # å½“å‰å·²çŸ¥çš„æœ€é•¿ç›¸ç­‰å‰åç¼€é•¿åº¦
    for right in range(1,m): # rightï¼šå½“å‰è€ƒå¯Ÿå­—ç¬¦ä¸‹æ ‡
        while left >0 and p[left] != p[right]: # leftä¸æ˜¯0ï¼Œå°±å›é€€åˆ°ä¸Šæ¬¡çš„å‰åç¼€
            left=next[left-1] # p[left-1]====p[right-1]ï¼Œä»ä¸Šä¸€ä¸ªå‰ä¸€ä½ç›¸åŒçš„æ•°å­—ç»§ç»­å¾€åçœ‹
        if p[left] == p[right]: # left=0å°±ä¸€ä¸ªä¸ªå¾€ä¸ŠåŠ 
            left+=1
        next[right] = left # è®°å½•å½“å‰æœ€é•¿ç›¸ç­‰å‰åç¼€é•¿åº¦
    return next

# KMPåŒ¹é…ç®—æ³•ï¼ŒTï¼šæ–‡æœ¬ä¸²ï¼Œpï¼šæ¨¡å¼ä¸²
def kmp(T:str,p:str) -> int:
    n,m=len(T),len(p)
    if m==0:
        return 0
    next=generateNext(p)
    j=0
    for i in range(n): 
        while j>0 and T[i]!=p[j]:
            # å›é€€jåˆ°next[j-1]
            j=next[j-1]
        if T[i] == p[j]:
            j+=1
        if j==m:
            return i-m+1
    return -1
```
