# 一、枚举算法
1、枚举算法：遍历所有可能解，找出满足需求的答案  

```python
from typing import List,Optional
class Solution:
    # 78 子集
    def subsets(self, nums: List[int]) -> List[List[int]]:
        ans=[]
        self.cur=[]
        def subset(nums,i,n):
            ans.append(self.cur[:])
            if i>=n:
                return
            for index in range(i,n):
                self.cur.append(nums[index])
                subset(nums,index+1,n)
                self.cur.pop()
        n=len(nums)
        subset(nums,0,n)
        return ans
    

        
        
            

s=Solution()            
nums=[1,2,3]
s.subsets(nums)        
                    
```

# 二、递归算法
1、递归算法：将复杂问题分解为与原问题结构相同的子问题，并通过重复求解这些子问题来获得最终答案的方法  
2、避免栈溢出：递归调用一次， 系统会分配一个栈帧，递归层数过深会导致栈溢出，改写为迭代算法显式栈模拟递归  
3、避免重复计算：使用哈希表记录已经计算的子问题结果

```python
from typing import List,Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    # 24 两两交换链表中的节点
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        dummy_head=ListNode(0)
        pre=dummy_head
        cur=head
        while cur:
            if cur.next:
                next=cur.next
                pre.next=next
                cur.next=next.next
                next.next=cur
                pre=cur
                cur=cur.next
            else:
                return dummy_head.next
        return dummy_head.next
    
    def swapPairs_Recursion(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        new_head=head.next
        head.next=self.swapPairs(new_head.next)
        new_head.next=head
        return new_head
    
    # 206 反转链表
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        new_head=self.reverseList(head.next)
        head.next.next=head
        head.next=None
        return new_head
    
    # 92 反转链表2
    def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:
        def reverse(head,n):
            if n==1:
                return head
            last=reverse(head.next,n-1)
            next=head.next.next
            head.next.next=head
            head.next=next
            return last
        if left==1:
            return reverse(head,right)
        head.next=self.reverseBetween(head.next,left-1,right-1)
        return head
    
    # 21 合并两个有序链表
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if list1 is None:
            return list2
        elif list2 is None:
            return list1
        elif list1.val < list2.val:
            list1.next=self.mergeTwoLists(list1.next,list2)
            return list1
        else:
            list2.next=self.mergeTwoLists(list1,list2.next)
            return list2
        
    # 70 爬楼梯
    def climbStairs(self, n: int) -> int:
        if n==0:
            return 1
        elif n<0:
            return 0
        return self.climbStairs(n-1)+self.climbStairs(n-2)
            
                
s=Solution()
s.climbStairs(3)         
                
        
        
        
```

# 三、分治算法
1、分治算法：把一个复杂问题拆分成多个相同或相似的子问题，递归分解，最后合并  

```python
from typing import List,Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    # 53 最大子数组和
    def maxSubArray(self, nums: List[int]) -> int:
        def max_sub_array(low,high):
            if low == high:
                return nums[low]
            mid = (low+high)//2
            leftMax=max_sub_array(low,mid)
            rightMax=max_sub_array(mid+1,high)
            
            total=0
            left_Total=-float("inf")
            for i in range(mid,low-1,-1):
                total+=nums[i]
                left_Total=max(left_Total,total)
            total=0
            right_Total=-float("inf")
            for i in range(mid+1,high+1):
                total+=nums[i]
                right_Total=max(right_Total,total)
            
            return max(leftMax,rightMax,left_Total+right_Total)
        return max_sub_array(0,len(nums)-1)
    
    # 23 合并K个升序链表
    def merge_sort(self,lists,left,right):
        if left==right:
            return lists[left]
        mid=(left+right)//2
        node_left=self.merge_sort(lists,left,mid)
        node_right=self.merge_sort(lists,mid+1,right)
        return self.merge(node_left,node_right)
    def merge(self,a,b):
        root=ListNode(-1)
        cur=root
        while a and b:
            if a.val<b.val:
                cur.next=a
                a=a.next
            else:
                cur.next=b
                b=b.next
            cur=cur.next
        if a:
            cur.next=a
        else:
            cur.next=b
        return root.next
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None
        size=len(lists)
        return self.merge_sort(lists,0,size-1)
    
    # 169 多数元素
    def majorityElement(self, nums: List[int]) -> int:
        def get_mode(low,high):
            if low==high:
                return nums[low]
            mid=(low+high)//2
            left_major=get_mode(low,mid)
            right_major=get_mode(mid+1,high)
            if left_major == right_major:
                return left_major
            else:
                left_cnt,right_cnt=0,0
                for i in range(low,high+1):
                    if nums[i] == left_major:
                        left_cnt+=1
                    elif nums[i] == right_major:
                        right_cnt+=1
                if left_cnt > right_cnt:
                    return left_major
                else:
                    return right_major
        return get_mode(0,len(nums)-1)
    
    # 14 最大公共前缀
    def longestCommonPrefix(self, strs: List[str]) -> str:
        def common(left,right):
            if left == right:
                return strs[left]
            mid=(left+right)//2
            left_common=common(left,mid)
            right_common=common(mid+1,right)
            idx=0
            while idx < min(len(left_common),len(right_common)):
                if left_common[idx] != right_common[idx]:
                    break
                else:
                    idx+=1
            return left_common[:idx]
        return common(0,len(strs)-1)
s=Solution()
strs = ["flower","flow","flight"]
s.longestCommonPrefix(strs)                
            
        
```

# 四、回溯算法
1、回溯算法：当发现当前路径无法满足题目要求或无法得到有效解时，算法会撤销上一步的选择（即「回溯」），返回到上一个决策点，尝试其他可能的路径

```python
from typing import List,Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    # 46 全排列
    def permute(self, nums: List[int]) -> List[List[int]]:
        ans=[]
        n=len(nums)
        path=[]
        def backtrack():
            if len(path)==n:
                ans.append(path[:])
                return
            for i in range(n):
                if nums[i] not in path:
                    path.append(nums[i])
                    backtrack()
                    path.pop()
        backtrack()
        return ans
    
    # 22 括号生成
    def generateParenthesis(self, n: int) -> List[str]:
        ans=[]
        path=[]
        def backtrack(idx1,idx2):
            if idx1 == n and idx2 == n:
                ans.append("".join(path))
                return
            elif idx1 < idx2 or idx1>n or idx2>n:
                return
            else:
                path.append("(")
                backtrack(idx1+1,idx2)
                path.pop()
                path.append(")")
                backtrack(idx1,idx2+1)
                path.pop()
        backtrack(0,0)
        return ans
    
    # 78 子集
    def subsets(self, nums: List[int]) -> List[List[int]]:
        ans=[]
        sets=[]
        n=len(nums)
        def subset(i):
            ans.append(sets[:])
            if i>=n:
                return
            for idx in range(i,n):
                sets.append(nums[idx])
                subset(idx+1)
                sets.pop()
        subset(0)
        return ans
    
    # 39 组合总和
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        ans=[]
        temp=[]
        n=len(candidates)
        def combination(idx,target):
            if target==0:
                ans.append(temp[:])
            elif target<0:
                return
            for i in range(idx,n):
                temp.append(candidates[i])
                combination(i,target-candidates[i])
                temp.pop()
        combination(0,target)
        return ans
    
    # 93 复原IP地址
    def restoreIpAddresses(self, s: str) -> List[str]:
        ans=[]
        temp=[]
        n=len(s)
        def IP(idx,size):
            if idx==n and size==4:
                ans.append(".".join(temp))
                return                
            if (n-idx) > (4-size)*3 or (n-idx) < 4-size:
                return
            if s[idx]=="0":
                temp.append("0")
                IP(idx+1,size+1)
                temp.pop()
            else:
                for i in range(1,min(4,n-idx+1)):            
                    last=s[idx:idx+i]
                    if int(last)<=255:
                        temp.append(last)
                        IP(idx+i,size+1)
                        temp.pop()
        IP(0,0)
        return ans
                
                
            
                    

        
        
s=Solution()
nums=[2,3]
candidates = [2,3]
str="101023"
s.restoreIpAddresses(str)
```

# 五、贪心算法
1、贪心算法：每一步都选择当前最优（看起来最好的）方案，期望通过一系列局部最优，最终获得全局最优解。

```python
from typing import List,Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    # 53 最大子数组合
    def maxSubArray(self, nums: List[int]) -> int:
        ans=-float("inf")
        cnt=0
        for num in nums:
            cnt+=num
            if cnt>ans:
                ans=cnt
            if cnt<0:
                cnt=0
        return ans
    
    # 122 买卖股票的最佳时机
    def maxProfit(self, prices: List[int]) -> int:
        ans=0
        for i in range(1,len(prices)):
            ans+=max(0,prices[i]-prices[i-1])
        return ans
                
            
    
```

# 六、位运算
常见的位运算符：

按位与（&）：两个对应位都为 1 时，结果位才为 1，否则为 0。  
按位或（|）：两个对应位中只要有一个为 1，结果位就为 1。  
按位异或（^）：两个对应位不同时，结果位为 1，相同则为 0。  
按位取反（~）：将每一位取反，即 0 变为 1，1 变为 0。  
左移（<<）：将二进制数向左移动指定的位数，右边空出的位用 0 填充。  
右移（>>）：将二进制数向右移动指定的位数，左边空出的位根据符号位填充。

```python
a=5 # 二进制表示为0101
b=3 # 二进制表示为0011
""" 按位与（&) """
print(a&b) # 0001——1
""" 按位或（|) """ 
print(a|b) # 0111——7
""" 按位异或（^) """
print(a^b) # 0110——6
""" 按位取反（~) """
print(~a) # 1010——-6（1开头代表负数，0开头代表正数）
""" 左移（<<) """
print(a<<2) # 010100——20=5*2**x
""" 右移（>>) """
print(a>>2) # 0001（01去掉）——1=[5/(2**2)]
```

```python
from typing import List,Optional
class Solution:
    # 78 子集
    def subsets(self, nums: List[int]) -> List[List[int]]:
        n=len(nums)
        sub_sets=[]
        for i in range(1 << n): # 1*2^n
            sub_set=[]
            for j in range(n): # 检查每一位
                if i >> j & 1:
                    sub_set.append(nums[j])
            sub_sets.append(sub_set)
        return sub_sets
    
    # 136 只出现一次的数字
    def singleNumber(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return nums[0]
        ans = 0
        for i in range(len(nums)):
            ans ^= nums[i]
        return ans
```
